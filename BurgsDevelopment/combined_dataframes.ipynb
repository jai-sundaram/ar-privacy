```python
# Hallways
hallway1 = pd.read_csv("../data/hallway1.csv")
hallway1 = hallway1.drop(0)
hallway2 = pd.read_csv("../data/hallway2.csv")
hallway2 = hallway2.drop(0)
hallway3 = pd.read_csv("../data/hallway3.csv")
hallway3 = hallway3.drop(0)
all_hallways = pd.concat([hallway1, hallway2], axis = 0)
all_hallways = pd.concat([all_hallways,hallway3], axis = 0)
all_hallways = all_hallways.drop(all_hallways.columns[0], axis = 1)
all_hallways = all_hallways.drop(all_hallways.columns[1], axis = 1)
```
